```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
```

Scraping Wikipedia file and storing it in a dataframe

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
df.head()
```

Dropping all entries where Borough is not assigned

```python
df.drop(df[df['Borough'] == 'Not assigned' ].index , inplace=True)
```

```python
df.head()
```

Checking for multiple entries of the same postal code

```python
df['Postal code'].value_counts()
```

Since all postal codes in the dataframe are unique I replace all the / with comma as prompted by the question

```python
df['Neighborhood']=df['Neighborhood'].str.replace('/',',')
```

```python
df.head()
```

Check for any Not assigned or blank neighborhood entries

```python
df['Neighborhood'].str.find('Not assigned') | df['Neighborhood'].str.find('')
```

Resetting index

```python
df.reset_index(drop=True,inplace = True)
```
